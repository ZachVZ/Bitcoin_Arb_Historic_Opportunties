# Crypto Arbitrage
This notebooks displays the phases and steps to analyze the historic arbitrage opportunities in Bitcoin and other cryptocurrencies between two exchanges: Bitstamp and Coinbase.
Below are the three phases that will be performed in this analysis: 
1. Collect the data
2. Prepare the data
3. Analyze the data 
4. Calculate the arbitrage profits

```python
## Required imports
import pandas as pd
from pathlib import Path
%matplotlib inline
```

# Phase 1. Collect the Data
This phase displays how to bring the datasets into your analysis.

### Step 1. The code below creates the dataframe name by importing the csv file from its path and replaces the current index with the date.

```python
df_bitstamp = pd.read_csv(
    Path("./Resources/bitstamp.csv"),
    index_col = "Timestamp",
    parse_dates = True,
    infer_datetime_format = True
)
```

### Step 2. The code below displays the first 5 rows of the dataframe imported to verify that the data was imported correctly.

```python
df_bitstamp.head()
```

### Step 3. The next two boxes bring in a second csv file named df_coinbase

```python
df_coinbase = pd.read_csv(
    Path("./Resources/coinbase.csv"),
    index_col="Timestamp",
    parse_dates=True,
    infer_datetime_format=True
)
```

```python
df_coinbase.head()
```

# Phase 2. Prepare the Data

### Step 1. The code below checks if there are any missing values and the following box will drop all NaNs or missing values in the DataFrame

```python
df_bitstamp.isnull().sum()
```

```python
df_bitstamp = df_bitstamp.dropna()
```

### Step 2. The next line will remove the $ sign in the close column by using the .loc and str.replace function.


```python
df_bitstamp.loc[:,"Close"] = df_bitstamp.loc[:,"Close"].str.replace("$","")
```

### Step 3. The following line will Convert the Close data type to a float by using the .astype("float")


```python
df_bitstamp.loc[:,"Close"] = df_bitstamp.loc[:,"Close"].astype("float")
```

### Step 4. Below will review the data for duplicate values, the second line of code is commeted outsince there are not lines to drop

```python
df_bitstamp.duplicated().sum()
#bitstamp = bitstamp.drop_duplicates()
```

### Step 5. The steps above will be repeated for the second dataframe df_coinbase

```python
df_coinbase = df_coinbase.dropna()
# Use the str.replace function to remove the dollar sign, $
df_coinbase.loc[:,"Close"] = df_coinbase.loc[:,"Close"].str.replace("$","")
# Convert the Close data type to a float
df_coinbase.loc[:,"Close"] = df_coinbase.loc[:,"Close"].astype("float")
# Review the data for duplicate values, and drop them if necessary
df_coinbase.duplicated().sum()
#drop duplicates
df_coinbase = df_coinbase.drop_duplicates()
```

```python
df_coinbase.duplicated().sum()
```

#### 4.1 For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one. Then use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame, where the spread is greater than zero.

```python
# For the three dates chosen in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early = df_bitstamp['Close'].loc["2018-01-28"] - df_coinbase['Close'].loc["2018-01-28"]
arbitrage_spread_middle = df_coinbase['Close'].loc["2018-02-17"] - df_bitstamp['Close'].loc["2018-02-17"]
arbitrage_spread_late = df_coinbase['Close'].loc["2018-03-30"] - df_bitstamp['Close'].loc["2018-03-30"]

# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
arbitrage_spread_early = arbitrage_spread_early[arbitrage_spread_early>0]
arbitrage_spread_early.describe()
```

```python
arbitrage_spread_middle = arbitrage_spread_middle[arbitrage_spread_middle>0]
arbitrage_spread_middle.describe()
```

```python
arbitrage_spread_late = arbitrage_spread_late[arbitrage_spread_late>0]
arbitrage_spread_late.describe()
```

#### 4.2 For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange youre buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early= arbitrage_spread_early / df_coinbase['Close'].loc["2018-01-28"]

# Review the spread return DataFrame
spread_return_early.describe()
```

```python
spread_return_middle = arbitrage_spread_middle / df_bitstamp['Close'].loc["2018-02-17"]
spread_return_middle.describe()
```

```python
spread_return_late = arbitrage_spread_late / df_bitstamp['Close'].loc["2018-03-30"]
spread_return_late.describe()
```

#### 4.3 For each of the three dates, narrow down the trading opportunities even further. To do so, determine the number of times the trades with positive returns exceed the 1% minimum threshold that are required to cover the costs.

```python
# For the date early in the dataset, determine the number of trades with positive returns 
# exceed the 1% minimum threshold (.01) to cover exchange costs
profitable_trades_early = spread_return_early[spread_return_early > .01]

# Review the first five profitable trades
profitable_trades_early.head()
```

```python
profitable_trades_middle = spread_return_middle[spread_return_middle > .01]
profitable_trades_middle.head()
```

```python
profitable_trades_late = spread_return_late[spread_return_late > .01]
profitable_trades_late.head()
```

#### 4.4 Generate the summary statistics of the spread returns that are greater than 1%. How do the average returns compare among the three dates?

```python
# For the date early in the dataset, generate the summary statistics for the profitable trades
# or you trades where the spread returns are are greater than 1%
profitable_trades_early.describe()
```

```python
profitable_trades_middle.describe()
```

```python
profitable_trades_late.describe()
```

#### 4.5 For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
# For the date early in the dataset, calculate the potential profit per trade in dollars 
# Multiply the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early *df_coinbase['Close'].loc["2018-01-28"]

# Drop any missing values from the profit DataFrame
profit_per_trade_early = profit_early.dropna()

# View the early profit DataFrame
profit_per_trade_early.head()
```

```python
profit_middle = profitable_trades_middle *df_coinbase['Close'].loc["2018-02-17"]

# Drop any missing values from the profit DataFrame
profit_per_trade_middle = profit_middle.dropna()

# View the middle profit DataFrame
profit_per_trade_middle.head()
```

```python
profit_late = profitable_trades_late *df_coinbase['Close'].loc["2018-03-30"]

# Drop any missing values from the profit DataFrame
profit_per_trade_late = profit_late.dropna()

# View the late profit DataFrame
profit_per_trade_late.head()
```

#### 4.7 Calculate the potential arbitrage profits that can be made on each day. To do so, sum the elements in the profit_per_trade DataFrame.

```python
# Calculate the sum of the potential profits for the early profit per trade DataFrame
profit_per_trade_early_sum = profit_per_trade_early.sum()
profit_per_trade_early_sum
```
